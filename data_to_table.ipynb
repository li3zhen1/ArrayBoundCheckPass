```python
import csv 
import pandas as pd


benchs = {
    "is": "large",
    "bfs": "large",
    "dither": "large",
    "jacobi-1d": "large",
    # "malloc_1d_array": "micro",
    # "static_1d_array": "micro",
    # "global_1d_array": "micro",
    "check_elimination": "micro",
    "check_modification": "micro",
}

def get_csv(bench: str, kind: str):
    df = pd.read_csv(f"stat_baseline/{bench}-{kind}.csv")
    df["kind"] = kind
    df["bench"] = bench
    df.drop(columns=["command", "stddev", "min", "max"], inplace=True)
    #reorder columns, put kind and bench in front
    cols = df.columns.tolist()
    cols = cols[-2:] + cols[:-2]
    df = df[cols]

    return df

all_dfs = pd.DataFrame()
    
for bench in benchs.keys():
    for k in ["original", "transformed"]:
        df = get_csv(bench, k)
        all_dfs = pd.concat([all_dfs, df])

for bench in benchs.keys():
    # mean col of transformed / mean col of original, 1 if this row's kind is original
    all_dfs[f"{bench}_speedup"] = all_dfs.groupby("bench")["mean"].transform(lambda x: x / x.iloc[0])
all_dfs
```
