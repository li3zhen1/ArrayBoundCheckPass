```python
import pandas as pd

benchs = {
    "is": "large",
    "bfs": "large",
    "dither": "large",
    "jacobi-1d": "large",
    "malloc_1d_array": "micro",
    "static_1d_array": "micro",
    "global_1d_array": "micro",
    "check_elimination": "micro",
    "check_modification": "micro",
}

def merge_count(bench):
    df = pd.read_csv(f"_stat_check_count/{bench}.txt", header=None, skipinitialspace = True)
    print(df)
    # group by the first column
    df = df.groupby(df.columns[1]).sum()
    # order row by column 1, ["After Insertion", "After Modification", "After Elimination", "After Loop Propagation"]
    df = df.reindex(["After Insertion", "After Modification", "After Elimination", "After Loop Propagation"])

    df.to_csv(f"_stat_check_count/{bench}.csv", header=False)
    # print(df)

for bench in benchs.keys():
    merge_count(bench)
```
